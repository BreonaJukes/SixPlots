```python
#this imports the matplotlib
import matplotlib.pyplot as plt

#these are lists of data for each slice of the pie
slices = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'working', 'playing']
#these are the colors for each slice
cols = ['c', 'm', 'r', 'b']

#this plots the pie chart
plt.pie(slices,
       labels = activities,
       colors = cols,
       startangle = 90,
       shadow = True,
       explode = (0, 0, 0, 0),
       autopct = '%1.1f%%')

#this displays the title and chart
plt.title('Pie Plot')
plt.show()
```
